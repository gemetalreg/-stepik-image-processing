```python
%matplotlib inline
```

```python
from skimage.io import imread, imshow, imsave
```

```python
img = imread('https://stepik.org/media/attachments/lesson/58181/tiger-color.png')
```

```python
imshow(img)
```

```python
from skimage import img_as_float
```

```python
img_f = img_as_float(img)
```

```python
r = img_f[:,:,0]
```

```python
imshow(r)
```

```python
g = img_f[:,:,1]
```

```python
imshow(g)
```

```python
b = img_f[:,:,2]
```

```python
imshow(b)
```

```python
from numpy import dstack
```

```python
img_combined = dstack((r,g,b))
```

```python
imshow(img_combined)
```

```python
avg_gray = (r+g+b)/3
```

```python
imshow(avg_gray)
```
