```python
from skimage.io import imread, imsave, imshow
from skimage import img_as_float, img_as_ubyte
from scipy.signal import convolve2d
import numpy as np
```

```python
s = '''https://stepik.org/media/attachments/lesson/61037/tiger-gray-small.png
'''
```

```python
img = imread(s)
```

```python
imshow(img)
```

```python
k = 5
kernel = np.ones((k,k), dtype=int)
```

```python
img = convolve2d(img, kernel, mode='valid') / k**2
```

```python
img = img.astype('uint8')
```

```python
imshow(img)
```

```python
imsave('out_img.png', img)
```
