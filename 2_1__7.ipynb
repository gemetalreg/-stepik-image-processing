```python
from skimage.io import imread, imshow, imsave
```

```python
img = imread('https://stepik.org/media/attachments/lesson/58180/tiger-gray.png')
```

```python
imshow(img)
```

```python
rimg = img.shape[0]
cimg = img.shape[1]

rows = 7
cols = 15
```

```python
r_start = (rimg - rows) // 2
r_end = r_start + rows
c_start = (cimg - cols) // 2
c_end = c_start + cols
```

```python
img[r_start:r_end, c_start:c_end] = [255, 192, 203]
```

```python
imsave('out_img.png')
```
