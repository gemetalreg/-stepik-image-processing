```python
from skimage.io import imread, imsave, imshow
from skimage import img_as_float, img_as_ubyte
import numpy as np
```

```python
img = imread('https://stepik.org/media/attachments/lesson/60611/landscape.png'
)
```

```python
imshow(img)
```

```python
values, bin_edge = np.histogram(img, bins = range(257))
```

```python
values
```

```python
cdf = np.cumsum(values)
```

```python
cdf
```

```python
cdf_min = cdf[cdf > 0].min()
```

```python
n_pix = img.size
```

```python
def f(x):
    return np.round((cdf[x] - cdf_min)/(n_pix - 1) * 255).astype('uint8')
```

```python
img_f = f(img)
```

```python
imshow(img_f)
```
